```python
#%%writefile data_gen.py
import os, glob, cv2, math, csv, tqdm, random
import numpy as np
import random
from PIL import Image
from PIL import ImageEnhance
import matplotlib.pyplot as plt

np.random.seed(42)
random.seed(42)

def aug_gauss(img, g):
    return cv2.GaussianBlur(img, (5,5), g)

def aug_con(img, c):
    img = img* c
    return np.minimum(img, 255).astype('int')

def aug_flip(img):
    return cv2.flip(img,1)

def img_read(path, w, h):
    img1 = cv2.resize(np.array(Image.open(path[0])),(w, h))
    img2 = cv2.resize(np.array(Image.open(path[1])),(w, h))
    return img1, img2
    
def aug_random(path, w, h):
    # path : list 2
    cond = random.randint(0,2)
    img1, img2 = img_read(path, w, h)
    if cond == 0:
        c = random.uniform(0.8,1.2)
        img1 = aug_con(img1, c)
        img2 = aug_con(img2, c)
    elif cond == 1:
        img1 = aug_flip(img1)
        img2 = aug_flip(img2)
    else :
        pass
        
    img_concat_f = np.concatenate((img1, img2),axis = 2)/255.
    img_concat_b = np.concatenate((img2, img1),axis = 2)/255.

    return img_concat_f, img_concat_b

def dataset_list_loader(kitti_path):
    dir_lists = sorted(os.listdir(kitti_path))
    total = []
    for i in dir_lists:
        paths = sorted(glob.glob(kitti_path + i + '/image_02/data/*.png'))
        x = [[paths[i], paths[i+1]] for i in range(len(paths)-1)]
        for j in x:
            total.append(j)          
    return total

def data_generator(total_img, batch):
    total = []
    random.shuffle(total_img)
    idx = 0
    while 1:   
        bat_img = []
        if idx > len(total_img) - batch:
            tmp_path = total_img[idx:]
            idx = 0
        else:
            tmp_path = total_img[idx:idx+16]
            idx = idx + batch

        for i in tmp_path:
            img1, img2 = aug_random(i,416,128)
            bat_img.append(img1)
            bat_img.append(img2)
        
        yield np.array(bat_img)
```

```python
def preprocess_image(image):
    # Convert from uint8 to float.
    return tf.image.convert_image_dtype(image, dtype=tf.float32)
```

```python
def augment_image_colorspace(image_seq):
    """Apply data augmentation to inputs."""
    # Randomly shift gamma.
    random_gamma = tf.random_uniform([], 0.8, 1.2)
    image_seq_aug = image_seq * random_gamma
    # Randomly shift brightness.
    random_brightness = tf.random_uniform([], 0.5, 2.0)
    image_seq_aug *= random_brightness
    # Saturate.
    image_seq_aug = tf.clip_by_value(image_seq_aug, 0, 1)
    return image_seq_aug
```

```python
kitti_path = '/data1/Kitti/usl/data_kitti/img/'
batch = 16
img_list = dataset_list_loader(kitti_path)
train = data_generator(img_list, batch)
```

```python
import os, glob, cv2, math, csv, tqdm, random
import numpy as np
import random
import tensorflow as tf
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def read_image(image_path):
    return np.array(Image.open(image_path))

def augment_random_val():
    return random.uniform(0.8,1.2), random.uniform(0.5, 2.0), [random.uniform(0.8,1.2), random.uniform(0.8,1.2), random.uniform(0.8,1.2)]
        
def augment_image_colorspace(image_t1, image_t):
    random_gamma, random_brightness, random_colors = augment_random_val()
    # Randomly shift gamma.
    image_t1_aug = image_t1**random_gamma
    image_t_aug = image_t**random_gamma
    # Randomly shift brightness. 
    image_t1_aug *= random_brightness
    image_t_aug *= random_brightness
    # Randomly shift color. 
    white = np.ones([image_t1.shape[0], image_t1.shape[1]])
    color_image = np.stack([white * random_colors[i] for i in range(3)], axis=2)
    image_t1_aug *= color_image
    image_t_aug *= color_image
    # Saturate.
    image_t1_aug = np.clip(image_t1_aug, 0, 1)
    image_t_aug = np.clip(image_t_aug, 0, 1)
    return image_t1_aug, image_t_aug
```
